#### Faça um programa em Python que calcule a média de notas de um aluno, cujos valores estão na lista  [10.0, 9.5, 9.0, 10]


```python
notas = [10.0, 9.5, 9.0, 10]
media = sum(notas) / len(notas)
print(f"A média das notas é: {media}")
```

#### Faça um programa em Python que, utilizando uma estrutura for, imprima a quantidade  de personagens que têm 43 anos e os seus nomes, conforme o dicionário abaixo: idade_personagens = { ‘Goku’: 43, ‘VeĀeta’: 48, ‘Chi-chi’: 43, ‘Bulma’: 47 }


```python
idade_personagens = {'Goku': 43, 'Vegeta': 48, 'Chi-chi': 43, 'Bulma': 47}
quantidade = 0
for personagem, idade in idade_personagens.items():
    if idade == 43:
        quantidade += 1
        print(f"{personagem} tem 43 anos.")
print(f"Quantidade de personagens com 43 anos: {quantidade}")
```

#### Crie uma função que recebe como  parâmetros uma taxa anual, um  período (em anos) e um capital  inicial e retorne o montante final da  aplicação. Dica: caso não for fornecido,  considere o período padrão de 1 ano


```python
def montante_final(capital_inicial, taxa_anual, periodo=1):
    return capital_inicial * (1 + taxa_anual)**periodo
```

#### Crie um programa que mostre em um  gráfico de linhas a evolução do  patrimônio de um investidor cujo  capital inicial é R$10 mil e fez uma  aplicação de 5% a.a. durante 30  anos. Dica: reutilize a função dos juros  compostos


```python
import matplotlib.pyplot as plt

def montante_final(capital_inicial, taxa_anual, periodo):
    return capital_inicial * (1 + taxa_anual)**periodo

anos = []
montantes = []

for ano in range(30):
    anos.append(ano + 1)
    montantes.append(montante_final(10000, 0.05, ano))

plt.plot(anos, montantes)
plt.xlabel('Anos')
plt.ylabel('Montante (R$)')
plt.title('Evolução do Patrimônio')
plt.show()
```
