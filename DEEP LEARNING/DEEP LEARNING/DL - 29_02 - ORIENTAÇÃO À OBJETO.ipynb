C = 50

t3 M3 = 50*1.1*1.1*1.1 = 50.1.1^3


```python
def juros_compostos(C, j, t=1):
  return C * (1+j) ** t
juros_compostos(500, 0.1, 3)
```

```python
investidor1 = {
    "nome": "João",
    "capital": 10000,
    "investimentos": {
        "CDB": 0.35,
        "acoes": 0.12,
        "LCA": 0.33,
        "LCI": 0.20
    }
}

investidor2 = {
    "nome": "Maria",
    "capital": 8000,
    "investimentos": {
        "CDB": 0.10,
        "acoes": 0.70,
        "FII": 0.20
    }
}

investidor1 = {
    "nome": "João",
    "capital": 10000,
    "investimentos": {
        "CDB": 0.35,
        "acoes": 0.12,
        "LCA": 0.33,
        "LCI": 0.20
    }
}

investidor3 = {
    "nome": "Gabi",
    "capital": 1000,
    "investimentos": {
        "CDB": 12.86,
        "acoes": 0.70,
    }
}

investidores = [investidor1, investidor2]
taxas = {
    'CDB': 0.01,
    'acoes':1,
    'LCI': 0.02,
    'LCA': 0.03,
    'FII': 0.09
}
```

```python
def montante_final(investidor, taxas, anos):
  montante = 0
  for investimento, percentual in investidor["investimentos"].items():
    montante += juros_compostos(investidor["capital"] * percentual, taxas[investimento], anos)
  return montante

investidor1_montante = montante_final(investidor1, taxas, 3)
investidor2_montante = montante_final(investidor2, taxas, 3)

print(f"Montante final do investidor 1: R${investidor1_montante:.2f}")
print(f"Montante final do investidor 2: R${investidor2_montante:.2f}")
```

#Orientação a objeto

```python
class Gato():
  #FUNÇÃO DENTRO DE CLASSE NÃO É FUNÇÃO É METÓDO
  def __init__(self, name, color):
    self.__nome = name #atributo privado
    self.cor = color #atributo público

  def miar(self):
    print(self.nome, " fez MIAAAAAAU")

  def get_nome(self, senha):
    if senha == "123":
      return self.__nome
    print("Senha inválida!")
    return None

  def set_nome(self, name):
    if name != "":
      self.__nome = name
```

```python
gatinho = Gato("Gerso", "Amarelo") #instanciando um objeto da classe gato
```

```python
gatinho2 = Gato("Adilson", "Rajado") #instanciando um objeto da classe gato
```

```python
print(gatinho.get_nome("123"))
print(gatinho2.get_nome("456"))
```
