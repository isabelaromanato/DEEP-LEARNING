```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
entradas = iris.data
saidas = iris.target
```

```python
# prompt: tamanho do df iris

import pandas as pd
df_iris = pd.DataFrame(entradas, columns=['comprimento_sepala', 'largura_sepala', 'comprimento_petala', 'largura_petala'])
print(df_iris.shape)
```

```python
redeneural = MLPClassifier(verbose=True, # a cada época vai imprimir na tela / aumenta o tempo de processamnto - nem sempre é recomendado
                           max_iter=10000, # quantas vezes quer que atualize o peso / épocas
                           tol=0.000001, # tolerância para "forçar" continuar treinando até a sexta casa decimal que estiver mudando
                           activation='relu', # relu = unidade linear retificada
                           #hideen_layer_sizes=(30,30,30), # tamanho da camada oculta, cada valor da tupla é uma camada e o núemro é a qtd de neurônio
                           learning_rate_init=0.0001) #
```

```python
redeneural.fit(entradas,saidas)
```

```python
redeneural.predict([[6.7,3.0,5.2,2.3]])
```
