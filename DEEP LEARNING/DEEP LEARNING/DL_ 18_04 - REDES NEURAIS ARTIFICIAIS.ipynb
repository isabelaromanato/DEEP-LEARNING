# Implementando um Perceptrón (rede neural de camada única)
                     

```python
import numpy as np
entradas = np.array([1, 2, 3])
pesos = np.array([0.1, 0.2, 0.3])

def somatorio(inputs, w):
   return inputs.dot(w)
   #opção completa
   '''total = 0
   for i in range(0 ,3):
   total += inputs[i]*w[i]
   return total'''
   #opção resumida
def degrau (u):
  if u >= 1:
    return 1
  return 0

u = somatorio(entradas,pesos)
output = degrau(u)
print("Soma:", u)
print("Saida:", output)
```

```python
def sigmoide(u, beta):
  return 1/(1 + np.exp(-beta*u))

entradas = np.array([-1.0, 5.0, 0.2])
pesos = np.array([0.11, -0.2, 4.0])

u = somatorio(entradas, pesos)
output = sigmoide(u , 1.25)
print("u=",u)
print("Saida=",output)
```
