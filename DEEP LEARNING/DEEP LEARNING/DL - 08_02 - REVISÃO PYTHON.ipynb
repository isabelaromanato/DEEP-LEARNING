```python
# correção professor 2
def equacao(a, b, c):
  if a == 0:
    if b!= 0:
      print("Equação do 1°grau")
      return -c/b
    else: #b == 0
      print("Não é possível calcular")
      return None
  else:
    delta = b**2 -4*a*c
    print(delta)
    if delta == 0:
      print('Delta zero')
      return -b/(2*a)
    else:
      if delta > 0:
        print('Delta positivo')
      else:
        print('Delta negativo')
      x1 = (-b + delta**0.5)/(2*a)
      x2 = (-b - delta**0.5)/(2*a)
      return (x1, x2)
a = float(input("Digite o valor de a:"))
b = float(input("Digite o valor de b:"))
c = float(input("Digite o valor de c:"))
equacao(a, b, c)
```

```python
# meu exercício - codigo para fazer o grafico da função acima

import matplotlib.pyplot as plt

def equacao(a, b, c):
  if a == 0:
    if b!= 0:
      print("Equação do 1°grau")
      return -c/b
    else: #b == 0
      print("Não é possível calcular")
      return None
  else:
    delta = b**2 -4*a*c
    print(delta)
    if delta == 0:
      print('Delta zero')
      return -b/(2*a)
    else:
      if delta > 0:
        print('Delta positivo')
      else:
        print('Delta negativo')
      x1 = (-b + delta**0.5)/(2*a)
      x2 = (-b - delta**0.5)/(2*a)
      return (x1, x2)

a = float(input("Digite o valor de a:"))
b = float(input("Digite o valor de b:"))
c = float(input("Digite o valor de c:"))

x1, x2 = equacao(a, b, c)

# Crie uma lista de valores x para o gráfico
x_values = [x for x in range(-10, 11)]

# Calcule os valores y correspondentes para cada valor x
y_values = [a * x**2 + b * x + c for x in x_values]

# Crie o gráfico
plt.plot(x_values, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfico da função")
plt.show()
```
