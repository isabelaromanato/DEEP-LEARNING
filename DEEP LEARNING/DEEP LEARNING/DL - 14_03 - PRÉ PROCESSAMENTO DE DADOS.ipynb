```python
import pandas as pd
import numpy as np
```

```python
dados = {
    "nome": ["Claudio", "José", "Flavia", "Maria"],
    "sexo": ["M", "M", "F", "F"],
    "salario": [2000.0, 6000.0, 10000.0, 20000.0],
    "bens": [0.0, 25000.0, 75000.0, 200.0],
    "score": [0, 600, 600, 1000]
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df = df.drop('nome', axis=1)
                # coluna - eixo 1
                # linha - eixo 0
```

```python
df
```

### Revisão Numpy

```python
exemplo = [10, 20, 30]
#mínimo
print(np.min(exemplo))
#máximo
print(np.max(exemplo))
#média
print(np.mean(exemplo))
#desvio padrão - o quão distante meus dados estão em relação a média
print(np.std(exemplo))
```

```python
def normalizar(x):
  return(x- np.min(x))/(np.max(x) - np.min(x))

df['salario'] = normalizar(df['salario'])
```

```python
df
```

```python
def padronizar (x):
  return(x - np.mean(x))/(np.std(x))
```

```python
df['bens'] = normalizar(df['bens'])
df['score'] = padronizar(df['score'])
```

```python
df
```

```python
exemplo = [10, 20, 30]
padronizar(exemplo)
```
