```python
class Produto():
  def __init__(self, id, nome, preco, descricao):
    self.__id = id
    self.nome = nome
    self.preco = preco
    self.descricao = descricao

  @property #(modifica o metodo para ser um getter)
  def id(self):
    return self.__id
  @id.setter #(modifica o método para ser um setter)
  def id(self, novo_id):
    self.__id = novo_id


ps5 = Produto(1, 'Playstation 5', 4000,"Melhor video game do mundo")
print(ps5.id)
ps5.id = 2
print(ps5.id)
```
